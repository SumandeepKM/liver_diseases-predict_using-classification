```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df['Albumin_and_Globulin_Ratio'].fillna(df['Albumin_and_Globulin_Ratio'].mean(), inplace = True)
```

```python
df.isnull().sum()
```

```python
df["Gender"].value_counts()
```

```python
df.rename(columns = {'Dataset': 'output'}, inplace = True)
df.head()
```

```python
df = pd.get_dummies(df)
df.head()
```

```python
df.drop('Gender_Female', axis = 1, inplace = True)
df.head()
```

```python
X = df.drop('output', axis = 1)
Y = df['output']
```

```python
X
```

```python
Y
```

```python
import seaborn as sns
sns.countplot(X['Gender_Male'])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2, random_state = 12)
```

```python
X_train.head()
```

```python
Y_train.head()
```

```python
X_test.head()
```

```python
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix, accuracy_score
classifier = SVC()
Svmodel = classifier.fit(X_train,Y_train)
Y_pred = Svmodel.predict(X_test)
cm = confusion_matrix(Y_pred,Y_test)
print(cm)
ac = accuracy_score(Y_pred, Y_test)
print(ac)
```

```python

from sklearn.tree import DecisionTreeClassifier
Dtmodel=DecisionTreeClassifier().fit(X_train,Y_train)
Y_pred = Dtmodel.predict(X_test)
cmd= confusion_matrix(Y_pred,Y_test)
print(cmd)
acd = accuracy_score(Y_pred, Y_test)
print(acd)
```

```python
from sklearn.ensemble import RandomForestClassifier
RC = RandomForestClassifier()
RCmodel = RC.fit(X_train, Y_train)
Y_pred = RCmodel.predict(X_test)
cmr = confusion_matrix(Y_pred,Y_test)
print(cm)
acr = accuracy_score(Y_pred, Y_test)
print(acr)
```

```python
import pickle
file = open("Random_forest.pkl",'wb')
pickle.dump(RCmodel, file)
model = pickle.load(open('Random_forest.pkl', 'rb'))
```
